## ðŸŒ„ Combine I et II : On peut faire the unique images into training_folder. 

```python
import sys
import pyautogui
import time
from PIL import ImageChops
from PIL import Image
import os

# Load all files
folder_name = input("Please input the ðŸ“• folder name , e.g. thread_main_game_win")
entries = os.listdir(folder_name)

# Remove .keep files...
try:
    entries.remove('.keep')
except:
    pass
try:
    entries.remove('.DS_Store')
except:
    pass

# Initial unigue_imag_list
unique_images_list = []

for any_file_name in entries:
    # Load 
    _tem_im = Image.open("%s/%s"%(folder_name,any_file_name)).convert('RGB')
    
    # The image must different with all unique_images_list
    diff_with_all_unique_imgs_flag = True
    
    # 1st run, empty list, save without comparing.
    if len(unique_images_list) == 0:
        unique_images_list.append(_tem_im)
    # Start to compare with unique_image_list
    else:
        for any_unique_im in unique_images_list:
            diff = ImageChops.difference(any_unique_im,_tem_im)
            diff_bbox = diff.getbbox()
            
            if diff_bbox is not None:
                continue       
            else:
                diff_with_all_unique_imgs_flag = False
                break
                
        # Check flag
        if diff_with_all_unique_imgs_flag == True:
            # print("Found one unique image")
            
            # append
            unique_images_list.append(_tem_im)
        else:
            continue
        
        

print("The final len of images = ",len(unique_images_list))
            
        
        
# Save:
saving_folder = input("Please set saving folder e.g. sample\n")
for index, any_im in enumerate(unique_images_list):
    any_im.save("%s/sample_%s.jpg"%(saving_folder,index),"JPEG")

print("Finihed !")

```

## Pandas DataFrame Format:
---

Table Example: Comme ca


|index | 0 | 1 
|:--- |:---|:---
|is_dix | True | False 
|___is_onze___ | False | True 
| is_cherry | True | False 

```python
index_col = []
zero_col = []
un_col   = []

def __name_conver_boolean(input_class_name, current_name):
    compare_value   = "%s"%input_class_name
    compare_value_2 = "not_%s"%input_class_name
    if current_name == compare_value:
        return True
    elif current_name == compare_value_2:
        return False
    else:
        print("Error: Unexpected current_name = %s"%current_name)
        

for any_norm in map_norm_reverse:
    cmd_ls = 'ls train_models/%s_dir/Train'%any_norm
    res = os.popen(cmd_ls).read()
    item_list = res.splitlines()
    
    # index_col faire ici:
    index_col.append("is_%s"%any_norm)
    
    for index , item_name in enumerate(item_list):
        print("(index, item_name) = (%d, %s)"%(index,item_name))
        
        if index == 0:
            tem_true_or_false = __name_conver_boolean(any_norm, item_name)
            zero_col.append(tem_true_or_false)  
        elif index == 1:
            tem_true_or_false = __name_conver_boolean(any_norm, item_name)
            un_col.append(tem_true_or_false)
        else:
            print("[Error]: Unexpected index %d"%index)
            

# Save to PD
index_name_map_df = pd.DataFrame({"class_name":index_col,'class_0':zero_col,'class_1':un_col})

# Set 'class_name' as inex
index_name_map_df = index_name_map_df.set_index('class_name')
print(index_name_map_df)

# Save to csv
index_name_map_df.to_csv('class_index_et_name.csv')
```
